```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("2.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
sampled =cv2.resize(img, (128,128), interpolation=cv2.INTER_NEAREST)
sampled_up = cv2.resize(sampled, (img.shape[1], img.shape[0]), interpolation=cv2.INTER_NEAREST)
def quantize_image(image, levels):
  div= 256//levels
  quantized = (image // div) * div + div //2
  return quantized

q_img_8 = quantize_image(img, 8)   # 8 levels per channel
q_img_16 = quantize_image(img, 16) # 16 levels per channel
q_img_32 = quantize_image(img, 32) # 32 levels per channel

fig, axs = plt.subplots(2, 3, figsize=(12,8))

axs[0,0].imshow(img); axs[0,0].set_title("Original")
axs[0,1].imshow(sampled_up); axs[0,1].set_title("Sampled (128x128)")
axs[0,2].imshow(q_img_8); axs[0,2].set_title("Quantized (8 levels)")

axs[1,0].imshow(q_img_16); axs[1,0].set_title("Quantized (16 levels)")
axs[1,1].imshow(q_img_32); axs[1,1].set_title("Quantized (32 levels)")
axs[1,2].axis("off")

plt.show()


```
